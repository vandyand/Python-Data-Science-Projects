```python
# Import modules
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create function to run an sql query and return a dataframe
def run_query(q):
    with sqlite3.connect('chinook.db') as conn:
        return pd.read_sql(q, conn)
```

```python
# Create function to execute sql commands
def run_command(c):
    with sqlite3.connect('chinook.db') as conn:
        conn.isolation_level = None
        conn.execute(c)
```

```python
# Create function to view the tables in sql database
def show_tables():
    q = '''SELECT name, type 
           FROM sqlite_master 
           WHERE type IN ("table","view");'''
    return run_query(q)
```

```python
show_tables()
```

# Sales by Genre
This section is intended to answer the question: "If we were to add another album to our store, what genre should it be?"

```python
# Rank genres by tracks sold
q = '''
select 
    g.name, 
    count(invoice_line_id) count,
    count(invoice_line_id)/(select cast(count(*) as float) from invoice_line) pct
from invoice_line il
inner join track t on t.track_id = il.track_id
inner join genre g on g.genre_id = t.genre_id
group by g.name
order by count desc
'''
genre_sales = run_query(q)
genre_sales
```

```python
# Visualize results using bar chart
genre_sales.set_index('name').plot.bar(y='pct')
plt.show()
```

Rock is by far the most prominent genre sold so it's recommended to add another rock album.

# Sales Support Agent Analysis
This section answers the questions, "Who is the best sales representative based on total sales?" and "Who is the best sales representative based on sales/day?"

```python
# See all employees from the employee table
q = '''
select *
from employee
'''
emp = run_query(q)
emp
```

```python
# Analyze the support reps based on how many records they've each sold.
q = '''
select 
    e.first_name||" "||e.last_name emp_name,
    e.hire_date,
    sum(i.total) total_sales
from employee e
inner join customer c on c.support_rep_id = e.employee_id
inner join invoice i on i.customer_id = c.customer_id
group by emp_name
'''
sales_agents = run_query(q)
sales_agents
```

```python
# Visualize results using bar chart
sales_agents.set_index('emp_name').plot.bar()
```

Jane Peacock is the sales agent with the most total sales. It's interesting that total sales of each employee corresponds pretty directly with their hire date. This makes sense because agents who have been at the company longer would make more sales naturally. Let's assess the efficiency of the sales agents.

```python
# Analize sales agents' sales per day
sales_agents['hire_date'] = pd.to_datetime(sales_agents['hire_date'],infer_datetime_format=True)
import datetime
sales_agents['days_employed'] = datetime.date(year=2018,month=3,day=5) - sales_agents['hire_date']
sales_agents['sales($)/day'] = sales_agents['total_sales']/sales_agents['days_employed'].dt.days
sales_agents
```

```python
# Visualize results
sales_agents.set_index('emp_name').plot.bar(y='sales($)/day')
```

So we see Steve Johnson is the most efficient sales agent the company has even though he has been at the company the least amount of time and has the lowest total sales. His sales per day is almost double that of Jane Peacocks' and Margaret Parks'!

# Customers by Country
This section answers the question "What country should we invest advertising into going forward?"
To assess this, let's find out which country has the highest value customers - that is, which contry has the highest money spent per customer.

```python
q = '''
with a as (
    select 
        case
            when(
                select count(*)
                from customer
                where country = c.country
            ) = 1
            then "Other"
            else c.country
        end as country,
        c.customer_id,
        i.*
    from customer c
    inner join invoice i on i.customer_id = c.customer_id
)

select
    country,
    count(distinct customer_id) num_customers,
    sum(total) total_sales,
    sum(total)/count(distinct customer_id) avg_customer_value,
    avg(total) avg_order_value
from a
group by country
order by num_customers desc
'''
countries = run_query(q)
countries
```

```python
countries[:-1].set_index('country').plot(kind='bar',subplots=True,layout=(2,2),legend=False,figsize=(12,12))
plt.show()
```

Czech Republic has by far the greatest average customer value. Keep in mind, however, that there are currently only two customers in the Czech Republic. This is not a statistically significant sample size to say if this trend will continue with any certainty.
